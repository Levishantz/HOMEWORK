```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the City and Ride Data

city_data = pd.read_csv("data/city_data.csv")
ride_data = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset

df_file = pd.merge(city_data, ride_data, on='city')
# Display the data table for preview
df_file.head()
```

```python
'Urban', 'Suburban', 'Rural'
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

def plot_stats(city_type, color_value):
    seperate_type = df_file['type']== city_type
    the_data_urban = df_file[seperate_type]
    

    average_urban = the_data_urban.groupby('city')['fare'].mean()

    urban_city_group=the_data_urban.groupby('city')['ride_id'].count()

    urban_driver = the_data_urban.groupby('city')['driver_count'].count()

    plt.scatter(urban_city_group, average_urban, s=urban_driver*5, alpha=0.5, color=color_value, edgecolor='black',
    linewidth=2, label=city_type)

    plt.legend(loc='upper right',title="City Types")
    
    plt.xlabel('Total Numbr of Rides (Per City)')
    plt.ylabel('Average Fare ($)')
    plt.title('Pyber Ride Sharing Data(2016)')
    
    seperate_type = df_file['type']== 'Suburban'
    
    plt.xlim(0, 40, 5)
    plt.ylim(15, 45, 4)   

    # 'wont run' plt.grid(color='b', likestyle='-', linewidth=2)
      
plot_stats('Urban', 'tomato')
plot_stats('Suburban', 'skyblue')
plot_stats('Rural', 'yellow')

# Save Figuren

#plt.savefig('levisfile')
```

## Total Fares by City Type

```python
# Calculate Type Percents

sum_of_fair =  df_file['fare'].sum()

Fare_grouping = round(df_file.groupby('type')['fare'].sum() / sum_of_fair * 100, 1)
Fare_grouping

explode = (0.0, 0.1, 0.1)

colors = [ 'yellow', 'lightskyblue','lightcoral']

Fare_grouping.plot.pie(shadow=True, explode=explode,
colors=colors, autopct='%1.1f%%', startangle=150)

plt.title('% of Total Fares by City Type')

# Build Pie Chart

# Save Figure
```

## Total Rides by City Type

```python
# Calculate Ride Percents

sum_of_type = df_file['type'].count()

percent_city = round(df_file['type'].value_counts()/sum_of_type*100, 1)

explode = (0.0, 0.1, 0.1)

colors = ['lightcoral', 'lightskyblue', 'yellow']

# Build Pie Chart

percent_city.plot.pie(shadow=True, explode=explode,
colors=colors, autopct='%1.1f%%', startangle=260)

plt.title('% of Total Rides by City Type')


# Save Figure
```
